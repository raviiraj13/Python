#Reshaping

```python
demo=np.arange(15)
print(demo)
```

```python
reshaped=demo.reshape((3,5))
print(reshaped)
```

```python
flat=reshaped.flatten()
print(flat)
```

```python
ravel=reshaped.ravel()
print(ravel)
```

```python
transpose=reshaped.T
print(transpose)
```
