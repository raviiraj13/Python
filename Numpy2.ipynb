#Sorting

```python
unsorted = np.array([5,4,26,4,56,7,5,3,5])
sorted=np.sort(unsorted)
print(sorted)
```

```python
unsorted_arr_2d = np.array([[4,7,2],[3,2,1],[9,4,3]])
print(unsorted_arr_2d)
print('\n')
print(np.sort(unsorted_arr_2d,axis = 0),"\n") #sorted column wise
print(np.sort(unsorted_arr_2d,axis = 1)) #sorted row wise
```

#Filter

```python
nums=np.arange(21,32)
print(nums,"\n")
even_nums=nums[nums % 2 == 0]
print(even_nums)
```

#Filter with mask

```python
mask = nums > 24
print(nums[mask])
```

#fancing indexing vs np.where()

```python
indices = [0,2,6]
print(nums[indices])
```

#Adding and removing data

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
combined = np.concatenate((arr1,arr2))
print(combined)
```

```python
a=np.array([1,2,3,4,5,6,7])
print(np.delete(a,[1,5]))
print(np.delete(a,1))
```

#Array Compatibility

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr3 =np.array(([7,8,9]))
print(arr1.shape == arr2.shape)
```

```python
arr=np.array([[1,2,3],[4,5,6]])
new_row=([7,8,9])
new_col=([[11],[12]])
new_array_row=np.vstack((arr,new_row))
print(arr,"\n")
print(new_array_row,"\n")
new_array_col=np.hstack((arr,new_col))
print(new_array_col)
```
